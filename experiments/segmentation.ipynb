Downloading and Loading the Oxford-IIIT Pet dataset.

```python
from fastai.vision.all import *

path = untar_data(URLs.PETS)

path.ls()

# List some image files
(path/"images").ls()[:5]
# List some mask files
(path/"annotations"/"trimaps").ls()[:5]
```

The model for building the SegmentationdataLoaders.

```python
import os

path = r"/root/.fastai/data/oxford-iiit-pet/"

# FastAI needs a function that: given an image path → returns the mask path
def get_mask(img_path):
    return os.path.join(path, "annotations", "trimaps", f"{img_path.stem}.png")

# Build the DataLoaders
dls = SegmentationDataLoaders.from_label_func(
    path,
    fnames=get_image_files(os.path.join(path, "images")),
    label_func=get_mask,
    codes=["background", "pet", "boundary"],  # names for classes 1, 2, 3
    item_tfms=Resize(256),
    batch_tfms=aug_transforms(),
    valid_pct=0.2,
    seed=67
)

# Visualize a batch — ALWAYS do this before training
dls.show_batch(max_n=6, figsize=(12, 6))
```
